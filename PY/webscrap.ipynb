```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://subslikescript.com/movie/Titanic-46435"
respons=requests.get(url).text
soup=BeautifulSoup(respons,"html.parser")
```

```python
print(soup)
```

```python
box=soup.find("article",class_="main-article")
```

```python
print(box)
```

```python
#Feacthing title
title=box.find("h1").text
print(title)
```

```python
#featching script
script=box.find("div",class_="full-script").text
print(script)
```

```python
f=open(f"{title}.txt","w",encoding="utf-8")
f.write(script)
f.close()
```

```python
#all movies link to movies file
url="https://subslikescript.com/movies"
respons=requests.get(url).text
soup=BeautifulSoup(respons,"html.parser")
box=soup.find("article",class_="main-article")
```

```python
links=[]
for i in box.findAll("a",href=True):
    links.append(i["href"])
print(links)
```

```python
for i in links:
    try:
        url="https://subslikescript.com"+i
        respons=requests.get(url).text
        soup=BeautifulSoup(respons,"html.parser")
        box=soup.find("article",class_="main-article")
        title=box.find("h1").text
        print(title)
        f=open(f"{title}.txt","w",encoding="utf-8")
        f.write(title)
        f.close()
    except:
        print(i)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://github.com/topics"
respons=requests.get(url).text
soup=BeautifulSoup(respons)
title=soup.findAll("p",class_="f3 lh-condensed mb-0 mt-1 Link--primary")
```

```python
len(title)
```

```python
description=soup.findAll("p",class_="f5 color-fg-muted mb-0 mt-1")
len(description)
```

```python
url_link=soup.findAll("a",class_="no-underline flex-1 d-flex flex-column")
len(url_link)
```

```python
d={'title':[],'description':[],'url_link':[]}
for i in range(0,30):
    d['title'].append(title[i].text)
    d['description'].append(description[i].text.strip())
    d['url_link'].append("https://github.com"+url_link[i]['href'])
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```

```python
df.to_csv("Lifox",index=None)
```

```python
url = "https://github.com/topics/3d"
respond=requests.get(url).text
soup = BeautifulSoup(respond)
box = soup.findAll("h3",class_="f3 color-fg-muted text-normal lh-condensed")
stars = soup.findAll("span",class_="Counter js-social-count")
dict = {"user":[],"title":[],"link":[],"star":[]}
for i in stars:
    if i.text[-1]=="k":
        dict["star"].append(float(i.text[:-1])*1000)
    else:
        dict["star"].append(float(i.text))
for i in box:
    l = i.findAll("a")
    dict["user"].append(l[0].text)
    dict["title"].append(l[1].text)
    dict["link"].append(l[1]["href"])
df = pd.DataFrame(dict)
df
```
