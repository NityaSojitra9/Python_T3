```python
import pandas as pd
```

```python
#Forcast
url=f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_Key}"
respond=requests.get(url)
cord=respond.json()
d={"date":[],"temp":[],"pressure":[],"humidity":[],"visibility":[],"desc":[]}
for i in cord["list"]:
    d["date"].append(i["dt_txt"])
    d["temp"].append(i["main"]["temp"])
    d["pressure"].append(i["main"]["pressure"])
    d["humidity"].append(i["main"]["humidity"])
    d["visibility"].append(i["visibility"])
    d["desc"].append(i["weather"][0]["description"])

df=pd.DataFrame(d)
print(df)
```

```python
#Air Pollution Forcast
url=f"http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={lat}&lon={lon}&appid={API_Key}"
respond=requests.get(url)
cord=respond.json()
d={"aqi":[],"so2":[],"no2":[],"pm10":[]}
for i in cord["list"]:
    d["aqi"].append(i["main"]["aqi"])
    d["so2"].append(i["components"]["so2"])
    d["no2"].append(i["components"]["no2"])
    d["pm10"].append(i["components"]["pm10"])
    
df=pd.DataFrame(d)
print(df)
```
